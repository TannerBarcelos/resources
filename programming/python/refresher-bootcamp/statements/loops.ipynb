### `for` loops
- Used to loop over iterable objects, items in a list, keys etc.

```python
lst = [1, 2, 3, 4, 5]

for element in lst:
    print(element)
    
for idx, val in enumerate(lst):
    print(f"Index: {idx} - Value: {val}")
    
name = 'bobby turner'
for char in name:
    print(char)

full_name = name.split()
for name in full_name:
    print(name)
    
lst2 = [1, 'a', 20, 44]

for i in lst2:
    if type(i) == int:
        print(i)
    else:
        print('Found a char')
        
        
di = {
    'name': 'tanner',
    'age': 27
}

# .keys() returns a list of the keys which we can iterate
for key in di.keys():
    print(key)
    
# .items() returns a list of the key/val pair tuples which we can iterate
for key, value in di.items():
    print(f"{key} => {value}")
    
# normal loop over dictionary also gives us each entry in the lookup table. We can use the iterator
# to index the hashmap, do logic on key or value, etc.
for k in di:
    entry = di[k]
    print(entry)
```

### `while` loops
- Similar to for loops but while loops only do something while some condition is true

```python
inp = ''
while(True):
    if inp == 'quit':
        break
    print(inp)
    inp = input('Enter something')
```

```python
some_list = [x for x in range(100)]

accum = 0
while(accum < len(some_list)):
    print(some_list[accum])
    accum+=1
```

```python
for i in range(1, 101):
    if i % 15 == 0:
        print('FizzBuzz')
    elif i % 3 == 0:
        print('Fizz')
    elif i % 5 == 0:
        print('Buzz')
```
